## The `print([data][, sep=' '][, end='\n'])` function

- All the parameters: data, sep and end are optional

````python
# print and empty line
print()
````

- `[data]` - This argument can be one or more data values to be printed, each separated by comma

````python
print(123)
````

````python
print(123, 8.5, "Hello, world!", True, "Good morning!!!", 834353, False)
````

````python
print(1,2,3,4,5)
````

````python
print(1, 2, 3, 4, 5)
````

````python
print(1,    2, 3,       4,        5)
````

- `[, sep=' ']`: sep is short for separator
- By default its value is space ' '
- So, in the output data values are separateed by space

````python
print(1, 2, 3, 4, 5, sep="...")
````

````python
print(1, 2, 3, 4, 5, sep="   ")
````

````python
print(1, 2, 3, 4, 5, sep="--")
````

````python
print(100,"hello", True, 183444.0, sep='...')
````

````python
print(1, 2, 3, 4, 5)
print(11, 12, 13, 14, 15)
print(21, 22, 23, 24, 25)
````

````python
print("hyd\namr\nche\ntri\nblr")
````

- `[, end='\n']` - the end prints its text after printing the last parameter
- Its default is is new line character (\n), but we can change that

````python
print(1, 2, 3, 4, 5, end="...")
print(11, 12, 13, 14, 15, end="---")
print(21, 22, 23, 24, 25)
````

````python
print(1, end='-')
print(2, end='_')
print(3, end='.')
print(4)
````

````python
x = 10
````

````python
y = input()
````

### The `int(value)` function

- If the value is a string then it shouldn't contain any non-numeric characters, then it converts it into an integer
- If it is not a string then it converts into an integer

````python
int(123)
````

````python
int(12.3)
````

````python
int(True)
````

````python
int('486')
````

````python
# int('48.6')
````

````python
# int('123a')
````

````python
x = int(input("Enter an integer: "))
y = int(input("Enter another integer: "))
z = x + y
print("Result:", z)
````

### The `float(value)` function

- Similar to the int() function to a slight difference
- If the value is a string and contians numeric and one decimal point then it can convert into a float, otherwise returns an error
- It can convert integer, floats, Booleans to floats

````python
float(123)
````

````python
float(12.3)
````

````python
float(False)
````

````python
float('123')
````

````python
float("12.3")
````

````python
# float("123.a")
````

````python
x = float(input("Enter an integer: "))
y = float(input("Enter another integer: "))
z = x + y
print("Result:", z)
````
