# Comprehensions

- Generating values in a list, sets, and dictionaries

## List comprehension

- Generating values in a list using a `for` loop and an expression
```
[exp for loop]
```

- For every iteration of the loop and values is added based on the `exp` to the list

````python
list_100 = [x for x in range(1, 101)]
print(list_100)
````

````python
list_of_squares = [x**2 for x in range(1, 101)]
print(list_of_squares)
````

````python
letters = [chr(i) for i in range(65, 91)]
print(letters)
````

````python
letters = [chr(i) for i in range(97, 123)]
print(letters)
````

````python
characters = [chr(i) for i in range(128, 256)]
print(characters)
````

````python
## Dictionary comprehension
````

````python
squares = {x:x**2 for x in range(1, 21)}
print(squares)
````

````python
num_to_char = {x:chr(x+64) for x in range(1, 27)}
print(num_to_char)
````

````python
cubes = {x:x**3 for x in range(1, 21)}
print(cubes)
````

## Set Comprehension

````python
{x * 5 for x in range(1, 11)}
````

````python
# 2 to 101 - arithmetic series
print({a for a in range(2, 102, 3)})
````
