### The `strptime()` method
- this method is of `datetime` construct
- The `p` in `strptime()` stands for `parsing`
##### Syntax
```
datetime.strptime(datetime_str, format_str)
```

#### format_str codes
```
%y - two-digit year number
%Y - four digit year number
%m - month number (1 - 12)
%d - day number (1 - 31)
%H - hour number (24-hour format)
%I - hour number (12-hour format)
%M - minute number
%S - seconds number
```

````python
dt = '19-06-2021'
dt = datetime.strptime(dt, '%d-%m-%Y')
print(dt)
````

````python
dt2 = '2020 09 - 17'
dt2 = datetime.strptime(dt2, '%Y %m - %d')
print(dt2)
````

````python
dt3 = "29th of 8th month in 2017 at 11 o clock, 30 minutes I visited doctor"
dt3 = datetime.strptime(dt3, "%dth of %mth month in %Y at %H o clock, %M minutes I visited doctor")
print(dt3)
````

````python
dt4 = "17th of 8th month, 2023 @ 17:30 time"
dt4 = datetime.strptime(dt4, "%dth of %mth month, %Y @ %H:%M time")
print(dt4)
````

### `strftime()` method
- To format datetime value according to the specified format
- The `f` in `strftime` stands for format
##### Syntax
```
datetime_obj.strftime(format_str)
```

#### format_str codes
```
%a - short weekday name (sun, mon, tue, ...)
%A - full weekday  name (Sunday, Monday, Tuesday, ...)
%b - short month name (Jan, Feb, Mar, ...)
%B - full month name (Janurary, Febrary, March, ...)
%y - two-digit year number
%Y - four digit year number
%m - month number (1 - 12)
%d - day number (1 - 31)
%H - hour number (24-hour format)
%I - hour number (12-hour format)
%M - minute number
%S - seconds number
%p - AM / PM specifier
```

````python
event_dt
````

````python
event_dt.strftime("(%a) %b %d, %Y")
````

````python
event_dt.strftime("%B %d, %Y (%A)")
````

````python
event_dt.strftime("%Y-%m-%d %I:%M %p")
````

````python
event_dt.strftime("Date: %B %d, %Y - Time: %H:%M:%S %p")
````

## WORKING WITH TIMESPANS (timedelta)

- The `timedelta` construct from `datetime` module helps to create time spans
- Timedelta measure time in days, seconds, and microseconds

##### Syntax
```
timedelta([days, ][seconds, ][microseconds, ][weeks, ][hours, ][minutes, ][seconds, ][milliseconds, ][microseconds,])
```

````python
from datetime import date, time, datetime, timedelta
````

````python
six_weeks = timedelta(weeks=6)
type(six_weeks)
````

````python
print(six_weeks)
````

````python
project_duration = timedelta(weeks=14, days=2)
start_date = date(2026, 6, 22)
end_date = start_date + project_duration
print(end_date)
````

````python
task_time = timedelta(hours=3, minutes=30)
now = datetime.now()
end_time = now + task_time
print(end_time)
````

````python
three_weeks = timedelta(weeks=3)
today = date.today()
three_weeks_ago = today - three_weeks
print(three_weeks_ago)
````

### 3 attributes and one method of timedelta object

````python
# microseconds attibute to get microseconds ignoreing days, seconds
td_obj.microseconds
````

````python
# total_seconds() to get all the seconds
td_obj.total_seconds()
````
