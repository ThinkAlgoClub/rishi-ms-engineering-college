# WORKING WITH DICTIONARIES

- A collection of unordered items, with no index
- Each item has a key and value pair, separated by colon (:)
- No duplicates of key values in a dictionary
- And key values are of immutable type
- Dictionary is mutable

##### Syntax
```
{key1:value1, key2:value2, key3:value3, key4:value4, ...}
```

### Creating dictionaries

````python
numbers = {0:'zero', 1:'one', 2:'two', 3:'three', 4:'four'}
print(numbers)
````

### Avoid KeyError using `get()` method
##### Syntax
```
get(key[, default_value])
```

- Provide key to get the associated value, if the is not there then it returns None
- Additionally, the optinal argument default_value is used to get the default_value then the key is not present

````python
countries = {
    "IN":"India", "US":"United States", "UK":"United Kingdom"
}
countries
````

````python
countries["IN"]
````

````python
countries["US"]
````

````python
# countries["DE"]   # KeyError
````

````python
countries.get("IN")
````

````python
countries.get("DE")
````

````python
countries.get("DE", "Unknown")
````

### The `for` loop to get keys of a dictionary

````python
countries["CA"] = "Canada"
countries["DE"] = "Germany"
countries["RU"] = "Russia"
countries["AU"] = "Australia"
countries["KSA"] = "Kingdom Of Saudi Arabia"
print(countries)
````

````python
for k in countries:
    print(k, end=" ")
````

### To get keys we can use `keys()` method

````python
for k in countries.keys():
    print(k, end=" ")
````

### To get values we can use `values()` method

````python
for v in countries.values():
    print(v, end=" | ")
````

### To get items (keys and values) use `items()` method
- Each item is tuple of key and value pair

````python
for item in countries.items():
    print(item)
````

````python
for k, v in countries.items():
    print("{:5} {:25}".format(k +":", v))
````

### Dictionary of dictionaries

````python
emp_details = {
    "1001":{"name":"kavya", "age":35, "city":"blr", "salary":38000.0},
    "1002":{"name":"anita", "age":38, "city":"che", "salary":75000.0},
    "1003":{"name":"ramya", "age":44, "city":"del", "salary":83000.0},
    "1004":{"name":"pravlika", "age":23, "city":"mum", "salary":48000.0},
    "1005":{"name":"sunita", "age":31, "city":"hyd", "salary":35000.0}    
}
````

````python
emp_details["1003"]
````

````python
emp_details["1002"]
````

````python
emp_details["1003"]["salary"]
````

````python
emp_details["1003"]["name"]
````

````python
emp_details["1003"]["city"]
````

````python
empID = input("Enter an employee ID: ")
print()
for key, value in emp_details[empID].items():
    print("{:12} {:>25}".format(key.upper() + ":", value))
````
