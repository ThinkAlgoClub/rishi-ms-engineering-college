# WORKING WITH DICTIONARIES

- A collection of unordered items, with no index
- Each item has a key and value pair, separated by colon (:)
- No duplicates of key values in a dictionary
- And key values are of immutable type
- Dictionary is mutable

##### Syntax
```
{key1:value1, key2:value2, key3:value3, key4:value4, ...}
```

### Creating dictionaries

````python
numbers = {0:'zero', 1:'one', 2:'two', 3:'three', 4:'four'}
print(numbers)
````

````python
# creating an empty dictionary
catalogue = {}
````

### Accessing values using keys
##### Syntax
```
dict_name[key]
```
- Use the key to get the associated value

````python
numbers
````

````python
numbers[0]
````

````python
numbers[2]
````

````python
numbers[4]
````

````python
courses = {1:"Computer Science", 2:"Information Technology",
           3:"Electrons and Communication", 4:"Civil Engineering",
          5:"Mechanical Engineering"}
course = int(input("Enter course ID: "))
courses[course]
````

### Adding items to a dictionary

##### Syntax
```
dict_name[newKey] = value
```

````python
numbers
````

````python
numbers[5] = 'five'
numbers[6] = 'six'
numbers[7] = 'seven'
numbers
````

### Updating a value for a specific key
##### Syntax
```
dict_name[key] = newValue
```

````python
print(courses)
````

````python
courses[2] = "Artificial Intelligence"
print(courses)
````
