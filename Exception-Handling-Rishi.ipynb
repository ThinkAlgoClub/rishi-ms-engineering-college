## The `try` and `except` construct

```python
try:
    # the code may cause some exception
    risky_code()
except SomeExcpetion:
    # when an exception occurs this code runs
    recovery_plan()
```

````python
try:
    value = int(input("Enter an integer: "))
    print("The integer input is:", value)
except ValueError:
    print("Not entered an integer value.")
    
print("Program continues to execute...")
````

**Note**: Exception terminates the program immediately. We need to handle excpetions with `except` clause.

````python
# ZeroDivisionError
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print("Result:", result)
except ZeroDivisionError:
    print("Cannot divide by zero")

print('Program continues...')
````

````python
# First exception is caught
try:
    value = int(input("Enter an integer: "))
    result = 100 / value
    print("Result:", result)
except ZeroDivisionError:
    print('Cannot divide by zero')
except ValueError:
    print("Invalid entry.")
    
print('Program continues...')
````

````python
# Using a parent class to catch all the errors in it
try:
    value = int(input("Enter an integer: "))
    result = 100 / value
    print("Result:", result)
except ArithmeticError:
    print('Some arithmetic error')
except ValueError:
    print("Invalid entry.")

    
print('Program continues...')
````

```python
# Use the the parent and child class names hierarchially
# Use from specific to general (child to parent)
try:
    risk_code()
except ZeroDivisionError:
    pass
except ArithmeticError:
    pass
except Exception:
    pass
```

```python
try:
    risky_code()
except FileNotFoundError:
    pass
except OSError:
    pass
except Exception:
    pass
```

````python
# grouping the error
try:
    value = int(input("Enter an integer: "))
    result = 100 / value
    print("Result:", result)
except (ValueError, ZeroDivisionError):
    print("Either value error or zero division error")

print("program continues...")
````

```python
# Avoid the bare except
try:
    risky_code()
except:
    recovery_plan()
```

### Full syntax for `try` and `except` statement

##### Syntax
```python
try:
    risky_code()
except ExceptionName as var:
    recovery_plan()...
else:
    # when there is no exception
    addtional_code()
finally:
    # always runs, with or without exception
    # used cleanup activities
    # like close files, etc.
    always_runs()
```

````python
try:
    f = open("courses.txt", 'r')
    content = f.read()
except FileNotFoundError:
    print("File doesn't exists")
except OSError:
    print("File exists but reading problem")
except Exception:
    print('Some unknown error')
else:
    print("File opened successfully")
finally:
    if f:
        f.close()
        print("file closed successfully")
````
