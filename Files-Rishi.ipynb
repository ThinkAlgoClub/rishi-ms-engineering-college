# WORKING WITH FILES

#### Two types of files:
1. Text files
2. Binary files

### Text files
- A file is a sequence of lines
- Each line contains characters and ends with new line (\n) character
- Examples: txt, csv, log, html, xml, json, etc.

### Binary files
- It is a sequence of bytes; may contain parts of text
- Its not a text file
- Examples: compressed files, database files, video, audio, image, etc

### File operations

1. Open the file - open() function
2. Read from the file or write to the file - read() / write() methods
3. Close the file - close() method

### The built-in `open(filename, mode)` function

- `filename` - is the filename and may include filepath
- `mode` - specify the mode to open a file
- This function returns a **file object**

#### file `modes`
1. r (read)
    - If the file doesn't exist it throws FileNotFoundError
    - If the file already exists then use the `read()` method to read the file contents
2. w (write)
    - If the file doesn't exist then it creates one
    - If it exists then it erases all the existing data, and we can write new data using `write()` method
3. a (append)
    - If the file doesn't exist then it creates one
    - If the file already exists we can append the data to the end of the file using `write()` method
4. b (binary)
    - In binary mode, 'rb' for reading and 'wb' for writing
**Note**: Opening a file in one mode will restrict the access to other modes until you close it and open it in the required / respected mode

### The `write` mode

````python
# a file that doesn't exist will be created in write mode
f = open("june18.txt", "w")
f.write("FILE TYPES WE HANDLE\n")
f.write("---------------------\n")
f.write("Text files\nCSV files\nBinary files\n")
f.close()
````

````python
# opening an existing file in write erases all the data
f = open("june18.txt", "w")
f.write("Three file operations:\n")
f.write("Open\nRead\nWrite")
f.close()
````

### The `append`` mode

````python
# opening a file that doesn't exist, creates a new file in append mode
f = open("rishi-ms.txt", "a")
f.write("COURSES\n")
f.write("Electronics\nComputer Science\nInformation Technology\nElectricity")
f.close()
````

````python
# open a file in append mode to add data to the end of the file
f = open("rishi-ms.txt", "a")
f.write("\nMechanical\nCivil\nPrinting\nArchitecture\n")
f.close()
````

### Using `for` loop to read each line

````python
f = open("rishi-ms.txt", "r")
for line in f:
    print(line, end="")
f.close()
````

### The `read` method
- This method reads entire file as one string

````python
file = open("rishi-ms.txt", "r")
print(file.read())
file.close()
````

### The `readlines()` method
- This method reads a file into a list, where each line is an item in the list

````python
file = open("rishi-ms.txt", "r")
print(file.readlines())
file.close()
````

### The `readline()` method
- This reads the next starting from the first line

````python
file = open("rishi-ms.txt", "r")
line1 = file.readline()
line2 = file.readline()
line3 = file.readline()
print(line1, end="")
print(line2, end="")
print(line3, end="")
file.close()
````

### Using `with` statement to open a file
- If we open a file using `with` statement then it automatically closes the file when the indented block ends
- The `as` keyword is used to give alias names to the objects

````python
# open a file to automatically close
with open("rishi-ms.txt", "r") as file:
    for line in file:
        print(line, end="")
````

## WORKING WITH `CSV` FILES

- CSV stands for comma-seprated valeus
- In a csv file each line/row is a record. Each row/record ends with a new line (\n) character
- Each record contains columns (fields), where each column is separated by comma (,)
- **NOTE:** In the open() function we need to use an additional parameter with value (newline="").

##### Example of csv file data
```
name,city,age,salary,dept
raju,hyd,35,28000.0,hr
kiran,blr,30,35000.0,sales
kumar,che,40,48000.0,admin
praveen,tri,25,18000.0,hr
```

### The `csv` module
- To work with csv files we need csv module
1. We need writer() function to create writer object, then using writerows() we write data to it
2. We need reader() function to create reader object

````python
student_data = [
    ["roll no", "math", "physics", "chemistry"],
    [1, 78, 79, 80],
    [2, 100, 99, 85],
    [3, 75, 75, 75],
    [4, 100, 100, 100],
    [5, 99, 98, 97]
]
````

````python
import csv
with open("rishi-marks.csv", "w", newline="") as file:
    w_obj = csv.writer(file)
    w_obj.writerows(student_data)
````

````python
with open("rishi-marks.csv", "r", newline="") as file:
    r_obj = csv.reader(file)
    for line in r_obj:
        print(line)
````

## WORKING WITH BINARY FILES

- Here, we use `pickle` module to work with binary files
- Has two functions:
    1. `dump()` for writing to a binary file
    2. `load()` for reading from a binary file

````python
student_data = [
    ["roll no", "math", "physics", "chemistry"],
    [1, 78, 79, 80],
    [2, 100, 99, 85],
    [3, 75, 75, 75],
    [4, 100, 100, 100],
    [5, 99, 98, 97]
]
````

````python
import pickle
with open("students.bin", "wb") as file:
    pickle.dump(student_data, file)
````

````python
with open("students.bin", "rb") as file:
    data = pickle.load(file)
    print(data)
````
