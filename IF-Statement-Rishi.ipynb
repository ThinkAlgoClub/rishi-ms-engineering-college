# `if` statement/construct

- Conditional branching construct
- The branching that is constructed using the `if` statement can be used to executed code from 'one' branch or 'no' branch

- *Syntax*:
```
if bool_exp:
    block
[elif bool_exp:
    block]...
[else:
    block]
```

```
if clause: required; only one if clause in a if statement
elif clause: optional; 0 to n
else clause: option; 0 to 1
```

- A block is a code that is placed under a clause
- All the statements inside block are equally indented

- bool_exp is used for `if` and `elif` clauses
- All the clauses are evaluated sequentially until a clause with bool_exp True is Found.
- Once the clause with bool_exp is True found then it executes the code from that clause and end the `if` statement
- If none of the bool_exp is True then it executes the statements from the `else` clause

````python
age = 17
if age >= 18:
   print("You can vote.")
   print("Voting is on 30th June.")
   print("See you at polling booth.")
print("Goodbye")
````

````python
age = 5
if age >= 18:
   print("You can vote.")
   print("Voting is on 30th June.")
   print("See you at polling booth.")
else:
    print("You can't vote now!")
    print("Wait for", 18-age, "years.")

# JUMPS HERE
````
