# Lambda functions

- Labmbda is a oneline function
- It creates a simple function with parameters and an expression
##### Syntax
```
lambda parameters: expression
```

````python
squares = lambda x: x**2
````

````python
squares(5)
````

````python
ab_square = lambda x, y: (x + y) **2
````

````python
ab_square(5, 3)
````

````python
odd_or_even = lambda x: x % 2 == 0
````

````python
odd_or_even(5)
````

````python
odd_or_even(10)
````
