# WORKING WITH LISTS

- A list is a collection of items
- A list contains a sequence of items, as the items are indexed by ordinal numbers
##### Syntax
```
[item1, item2, item3, ...]
```

##### Indexing
- Every item has an index number
- The first item's index is 0, second item's index is 1, third item's index is 2, so on and so forth.
- The last item's index is -1, second last item's index is -2, third last item's index number is -3, so on and so forth.

##### Function
```

```

- Method manipulates the object

### The `in` keyword
- The `in` is a membership keyword
##### Syntax:
```
item in list
```
- If the item is in the list it returns True; otherwise False

````python
states = ['ts', 'ap', 'mp', 'kl' 'tn', 'ka', 'mh', 'wb']
````

````python
'ap' in states
````

````python
'dl' in states
````

````python
if 'dl' in states:
    pass   # do nothing
else:
    states.append('dl')

print(states)
````

### The `reversed(list)` function
- Reverses the order of the items during iteration

````python
for state in reversed(states):
    print(state, end=" ")
````

### The `enumerate(list)` function
- Iterative function, which returns each item along with its index number

````python
for i, state in enumerate(states):
    print(str(i+1) + ".", state)
````
