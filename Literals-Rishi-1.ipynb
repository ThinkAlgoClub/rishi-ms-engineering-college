# Literals

## Literals are those values
- Which don't need interpretation, they are obvious, clear, and raw
- Four types of literals
    - Integers (without decimal point or fractional part)
    - Floating-point numbers (with decimal point for fractional part)
    - String / text (sequence of characters)
    - Boolean values: True and False

````python
100.1 + 100.1 + 100.1
````

````python
100 + 50
````

````python
100 / 25
````

### Arithmetic operators
```
-> + addition
-> - subtraction
-> * multiplication
-> / division
-> // integer (floor) division
-> % modulus (remainder)
-> ** exponent (power)
```

````python
100 / 25
````

````python
27 / 4
````

````python
27 // 4  # quotient
````

````python
27 % 4  # remainder
````

````python
2 ** 1000
````

### Comparison Operators
```
-> == (equal to)
-> != (not equal to)
-> > (greater than)
-> < (less than)
-> >= (greater than or equal to)
-> <= (less than or equal to)
```

````python
100 == 100
````

````python
100 == 50
````

````python
50 != 75
````

````python
50 != 50
````

#### The `and` operator
- Combines two Boolean expressions
- **Syntax**:
```
bool_exp1 and bool_exp2
```

- When both the boolean expression are True it returns True; otherwise False

````python
True and True
````

````python
True and False
````

````python
False and True
````

````python
False and False
````

````python
True and True and True and True and True and True
````

````python
True and True and True and True and False and True
````

### The `or` operator
- It combines two boolean expression
- **Syntax**
```
bool_exp1 or bool_exp2
```

- When at least one bool_exp is True the result is True; otherwise False

````python
True or True
````

````python
True or False
````

````python
False or True
````

````python
False or False
````

````python
True or True or True or False or True or False or True or False or False
````
