# Loops

- A loop executes a block (code) repeatedly, and each repeation (execution) is called an iteration.

### Two types of loops
- Collection (countable) loop - for
- Conditional loop - while

### The `for` loop

##### **Syntax**
```
for <var> in <collection>:
    block
```

- `<collection>` is the collection of items for characters
- The loop iterates for each item in the collection.
- That means, for each item the the block executes once
- If there are n items in the collection the loop iterates for n times (block executes n times)
- In each iteration, the an item is assigned to the variable and then executes the block
- `block` is indented set of statements 

````python
for i in [10, -8, 55, 79]:
    print("Hello, world!!!")
    print("---------------")
````

````python
for e in ["1003", "Raju", 42, 85000.0, 'Hyd', True]:
    print("->", e)
````

````python
for l in 'hello':
    print(l)
````

````python
total = 0
for i in [1, 2, 3, 4, 5]:
    total = total + i
print(total)
````

### The `range()` ITERATOR FUNCTION

#### **Syntax**
```
range([start=0, ] stop [, step=1] )
```

- `[start=0]`: optional; its default value is 0
- stop: required
- `[step=1]` : optional; its default value is 1
- The `range()` is an iterator function which returns next value for each iteration. Strating from `start` value till `stop` value, by continuously adding `step` to `start` value till it reaches `stop` value. However, `stop` is exclusive.

#### range(stop)
- start = 0; step = 1; stop > 0

```
range(5)    # 0, 1, 2, 3, 4
range(8)    # 0, 1, 2, 3, 4, 5, 6, 7
range(3)    # 0, 1, 2
```

````python
for i in range(5):
    print(i, end=' ')
````

````python
for i in range(8):
    print(i, end=' ')
````

````python
for i in range(3):
    print(i, end=' ')
````

#### range(start, stop)
step = 1; start < stop

```
range(1, 5)     # 1, 2, 3, 4
range(8, 14)    # 8, 9, 10, 11, 12, 13
range(-4, 4)    # -4, -3, -2, -1, 0, 1, 2, 3
range(3, -3)    # WRONG
```

````python
for i in range(1, 5):
    print(i, end=' ')
````

````python
for i in range(8, 14):
    print(i, end=' ')
````

````python
for i in range(-4, 4):
    print(i, end=' ')
````

````python
for i in range(3, -3):
    print(i, end=' ')
````

#### range(start, stop, step)
- step is +ve; start < stop
- step is -ve; start > stop

```
range(3, 30, 6)       # 3, 9, 15, 21, 27
range(80, 160, 20)    # 80, 100, 120, 140
range(100, -100, -30) # 100, 70, 40, 10, -20, -50, -80
range(-500, 500, 200) # -500, -300, -100, 100, 300
range(10, -10, 5)     # WRONG
range(-10, 10, -5)    # WRONG
```

````python
for i in range(3, 30, 6):
    print(i, end=' ')
````

````python
for i in range(80, 160, 20):
    print(i, end=' ')
````

````python
for i in range(100, -100, -30):
    print(i, end=' ')
````

````python
for i in range(-500, 500, 200):
    print(i, end=' ')
````

````python
total = 0
for i in range(1, 6):
    total = total + i
print(total)
````

````python
total = 0
for i in range(1, 10001):
    total = total + i
print(total)
````

## The `while` loop

- The `while` is a conditional loop
- This loop iterates based on a condition
- We can code indefinite loop

#### **Syntax**
```
while bool_exp:
    block
```

- The loop iterates as long as the bool_exp is True; once the bool_exp becomes False the loop ends

````python
# infinite loops
while True:
    print("Good afternoon!")

while 1 < 5:
    print("Hello, world!")
````

````python
# loops that never run
while False:
    print("Goodbye!!!")

    
while 1 > 5:
    print("Hello, Pythonists!")
````

````python
# printing 0 to 4; start = 0, stop = 5, step = 1
for i in range(5):
    print(i, end=' ')
````

````python
i = 0     # start
while i < 5:   # stop
    print(i, end=' ')
    i += 1    # step
````

````python
total = 0
i = 1
while i < 6:
    total += i
    i += 1
print(total)
````

### The `break` statement

````python
emps = ['ramya', 'kumari', 'sravani', 'kavya', 'jaya', 'anitha']

for e in emps:
    print(e)
    if e == 'sravani':
        break
````

````python
numbers = [85, 29, 54, 61, 56, 81, 99, 12]
for n in numbers:
    print(n, end=' ')
    if n % 7 == 0:
        break
````
