### 12. `find(substring[, start_position])` method
- Returns the starting index of the first occurance of the substring
- Optionally we can provide staart_position index number, to start searching form the substring
- If the substring is not in the larger string then it returns -1

````python
email = "kasala.swathi@gmail.com"
````

````python
at_index = email.find("@")
at_index
````

````python
dot_index = email.find(".", at_index + 1)
dot_index
````

````python
if at_index > -1 and at_index < dot_index:
    print("Valid email address")
````

### 18. `format(data_item(s))` method
- It used to format data items, and the format specification is inside the string ('')

##### Syntax
```
"...{:format_specification}...".format(data_item)
```

#### `format_specification` has five components
```
[alignment][width][comma][.deci_places][type_code]
```
---
```
[alignment] - By default text is left-aligned and numbers are right aligned.
            - '<' for left-alignment
            - '>' for right-alignment
            - '^' for center alignment
[width]     - The width for the data-item
[comma]     - thousand separator in numbers
[.deci_place] - specify number of decimal places
[type_cdoe] - 'd' for integers, 'f' for floats, and 'e' for scientific notation
```

````python
"{:20}".format("hello")
````

````python
"{:15}".format(12345)
````

````python
"{:>20}".format("hello")
````

````python
"{:^20}".format("hello")
````

````python
"{:<20}".format(1234)
````

````python
num = 123456789012345
"{:25,}".format(num)
````

````python
x = 123456789.23872454
````

````python
"{:,.2f}".format(x)
````

```
Product_name      Units        Price
product1              1     18000.00
product2            103       345.00
product3             15      1400.00
...                 ...          ...
```

````python
print("{:20} {:>12} {:>15}".format("PRODUCT NAME", "UNITS", 'PRICE'))
print("{:20} {:>12} {:>15}".format("------------", "-----", '-----'))
print("{:20} {:12} {:15.3f}".format("HP Ryzen 7", 20, 78000.0))
print("{:20} {:12} {:15.3f}".format("Epson Projector", 4, 47500.0))
print("{:20} {:12} {:15.3f}".format("HP Inkjet Printer", 7, 19500.0))
print("{:20} {:12} {:15.3f}".format("Logitech Webcam", 12, 3000.0))
````
