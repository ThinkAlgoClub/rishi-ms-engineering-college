# WORKING WITH TUPLES

- A collection of sequentional items
    1. Every item is indexed just like with lists
    2. Positive indexing to refer items from first to last
    3. Negative indexing to refer items from last to first, just like a list 
- A tuple is immutable

##### Syntax
```
(item1, item2, item3, item4, item5, ...)
```

````python
# Example
week_days = ('Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat')
````

#### Using indexing

````python
week_days[0]
````

````python
week_days[1]
````

````python
week_days[-1]
````

````python
week_days[-2]
````

````python
# week_days.append('Jan')  # AttributeError
````

````python
sorted(week_days)
````

````python
len(week_days)
````

````python
for day in reversed(week_days):
    print(day, end=" | ")
````

````python
for i, day in enumerate(week_days):
    print(str(1 + i) + ".", day)
````

### Slicing

##### Syntax
```
tuple[start:end:step]
list[start:end:step]
```

````python
week_days
````

````python
week_days[0:3]
````

````python
week_days[:3]
````

````python
week_days[::2]
````

````python
week_days[1::2]
````

````python
week_days[::3]
````

````python
week_days[::-1]
````
