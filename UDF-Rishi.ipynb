# User-Defined Functions (UDFs)

- Avoid repetition (Functions can be reused)
- You can make a program manageable (by divide into function chunks)

#### **Syntax for defining a function**
```
def func_name([args]):
    block
```

#### **Syntax for calling a function**
```
func_name([args])
```

- `def` - to define/create a function
- `func_name` - to name a function with the same naming rules that apply to variables
- `[args]` - optional; args are inputs to a function
- `block` - the indented block as part of function
- Functions don't excute their block unless you 'call the function'

````python
# function definition; without arguments
def welcome():
    print("Hello, World!")
    print("I am learning Python to tame it.")
    print("Join with me in this journey")

# function call
welcome()
````

````python
# function definition; one argument
def welcome(name):
    print("Hello,", name)
    print("I am learning Python to tame it.")
    print("Join with me in this journey")

welcome("Prasanna")
````

````python
# function definition; two arguments
def welcome(name, age):
    print("Hello,", name)
    print("Age:", age)

welcome("Aparna", 20)
````

````python
# function definition; three parameters
def product(name, price, discountPercent):
    print("Name", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price* (1 - discountPercent/100))

product('Samung A22', 22000.0, 5)
````

### Creating an optional parameter with a default value

````python
# two are required and one is optional parameter
# by setting a value to a parameters in the first line of the function
# definition the parameter becomes optional
# to overwrite a default value pass a value to the parameter
def product(name, price, discountPercent=2):
    print("Name", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price* (1 - discountPercent/100))

product('BenQ', 30000.0, 10)
````

- **Rule**: First code required parameters then code optional parameter

### Named arguments to deal with positional parameters

````python
def product(name, price, discountPercent=2):
    print("Name", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price* (1 - discountPercent/100))

# named parameters will help us to bypass positional aspect to those
product(price=40000.0, name='HP Ryzen 5')
````

### The `return` statement

- The `return` statement returns a value(s) or an object to the calling statement
- Which means, the calling statement is replaced by the returned value/object
- The `return` statement ends the function


````python
def sample(x):
    x = 3 * x
    print(x)

x = 5
sample(x)
````

````python
def sample(x):
    x = 3 * x
    return x

x = 5
sample(x)
````

````python
def sample(x):
    x = 3 * x
    return x
    print("Good morning!")
    print("Good evening!")

x = 5
res = sample(x)
print("Result:", res)
````

### Local and Gloabl variables

#### Global variables
- These variables are defined outside of any function
- Global variable is accessbile everywhere such as outside of any function as readable and writable varaibale; inside a function its accessable as read-only variable (only readable inside a function)
- We can make a global variable writable inside a function using `global` keyword, which is not recommended.

#### Local variables
- These variables are defined / created inside a function
- They are only accessible inside the function in which they are defined

**Note**: If local and global variables have same name, then inside function it is local variable that has precedence

````python
def sample():
    print(x)   # global variables
    
x = 25  # global variable
sample()
````

````python
def sample():
    x = 25    # local variable
    print(x)  # local variable

x = 100    # Global variable
sample()
print(x)   # Global variable
````

````python
def sample(x):
    y = 3 * x   # x and y are local
    print(y)
    
x = 25      # global
sample(x)   # global
print(x)    # global
````

````python
def sample():
    global x      # global as writable; not recommended
    x = x * 10    # updating global variable
    print(x)
    
x = 12     # global
sample()
print(x)   # global
````

````python
TAX = 0.25

def tax_amount(salary):
    return salary * TAX

def paid_salary(salary):
    return salary * ( 1 - TAX )

salary = 50000
print("Tax:", tax_amount(salary))
print("Take home:", paid_salary(salary))
````
