# WORKING WITH MODULES

- A moudle is a python file
- If name of the file is `temperature.py` then module name is `temperature`

- To use a module we need to import the module:
```
import module_name
```

import temperature

````python
import temperature
````

##### Using functions from a module
```
module_name.func_name([args])
```

temperature.to_celsius(100)

````python
temperature.to_celsius(100)
````

````python
temperature.to_fahrenheit(40)
````

#### Use short alias names for imported modules using the keyword `as`

````python
import temperature as t
t.to_celsius(88)
````

#### Directly load functions from a module; not recommended

````python
from temperature import to_celsius, to_fahrenheit   # not recommended
````

````python
to_celsius(200)
````

## Standard library modules

```
1. math - contains functions for mathematical operations
2. random - contains functions to generate random values
3. csv - contains functions to work with csv files
4. json - contains functions to work with json files
5. datetime - contains functions to work with date, time, and datetime
6. pickle - used for serialization and de-serialization of data
7. string - to work with strings
```
