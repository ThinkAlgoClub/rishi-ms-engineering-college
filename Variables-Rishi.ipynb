### Rules for naming variables
- Use descriptive names
- A variable name can contain letters (upper and lowercase), numbers and underscore character
- A variable shouldn't begin with a number
- Variable names are case-sensitive
- Don't use keywords for variable names

````python
sales_2026_may = 8500000
````

````python
roi = 0.12
````

````python
emp_id = '1003'
````
